```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
students = pd.read_csv('students.csv')

# Print first few rows of data
print(students.head())

# Print summary statistics for all columns
print(students.describe(include='all'))
area=students.address.value_counts()
print(area)
#students live in mostly in the urban areas

# Calculate mean
print('grade mean', students.math_grade.mean())
# Calculate median
print('grade median',students.math_grade.median())
# Calculate mode
print('grade mode', students.math_grade.mode())
# Calculate range
print('grade range', students.math_grade.max()-students.math_grade.min())
# Calculate standard deviation
print('grade sd', students.math_grade.std())
# Calculate MAD, mad() is deprecated
print('grade mean absolute deviation',abs(students["math_grade"]-students["math_grade"].mean()).mean())
```

```python
# Create a histogram of math grades

sns.histplot(x =students.math_grade, data = students.math_grade)
plt.show()
plt.clf()
#This ensures that the plots dont get layered on top of each other.

# Create a box plot of math grades

sns.boxplot(x =students.math_grade, data = students.math_grade)
plt.show()
plt.clf()

# Calculate number of students with mothers in each job category

print('mom job type frequencies', students.Mjob.value_counts())

# Calculate proportion of students with mothers in each job category

print('mom job type proportions', students.Mjob.value_counts(normalize=True))

# Create bar chart of Mjob
sns.countplot(x=students.Mjob,data=students)
plt.show()
plt.clf()
# Create pie chart of Mjob
students.Mjob.value_counts().plot.pie()
plt.show()
plt.clf()

# Create bar chart of Fjob
sns.countplot(x=students.Fjob,data=students)
plt.show()
plt.clf()


# Create a histogram of absences

sns.histplot(x =students.absences, data = students.math_grade)
plt.show()
plt.clf()
```

```python
# Frequency contingency proportions
obs_freq = pd.crosstab(students.Mjob,students.Fjob)
print(obs_freq/len(students.Mjob))
```
